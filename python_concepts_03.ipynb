```python
import matplotlib.pyplot as plt
import pandas as pd
import re
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(newsgroups.data, newsgroups.target, random_state=42)
```

```python
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)
```

```python
y_pred = classifier.predict(X_test_vec)
```

```python
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
print(f"F1 Score: {f1_score}")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the 311 Service Requests data
na_values = ['NO CLUE', 'N/A', '0']
requests = pd.read_csv('/content/311-service-requests.csv', na_values=na_values, dtype={'Incident Zip': str}, low_memory=False)

def fix_zip_codes(zips):
    # Truncate everything to length 5
    zips = zips.str.slice(0, 5)

    # Set '00000' zip codes to nan
    zero_zips = zips == '00000'
    zips[zero_zips] = np.nan

    return zips

# Apply the fix_zip_codes function to 'Incident Zip'
requests['Incident Zip'] = fix_zip_codes(requests['Incident Zip'])

# Convert 'Incident Zip' to string before sorting
requests['Incident Zip'] = requests['Incident Zip'].astype(str)

# Display unique values after cleaning
unique_zips = requests['Incident Zip'].unique()
unique_zips.sort()
print(unique_zips)
```

```python
import nltk

# Download the punkt resource
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, RegexpParser

# Sample text from Moby Dick Chapter 133
text = """That night, in the mid-watch, when the old man—as his wont at intervals—stepped forth from the scuttle in which he leaned, and went to his pivot-hole, he suddenly thrust out his face fiercely, snuffing up the sea air as a sagacious ship’s dog will, in drawing nigh to some barbarous isle. He declared that a whale must be near. Soon that peculiar odor, sometimes to a great distance given forth by the living sperm whale, was palpable to all the watch; nor was any mariner surprised when, after inspecting the compass, and then the dog-vane, and then ascertaining the precise bearing of the odor as nearly as possible, Ahab rapidly ordered the ship’s course to be slightly altered, and the sail to be shortened."""

# Tokenize the text
words = word_tokenize(text)
print("Tokenized Text:")
print(words)

# Filter stop words
stop_words = set(stopwords.words("english"))
filtered_words = [word for word in words if word.lower() not in stop_words]
print("\nText after removing stop words:")
print(filtered_words)

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("\nStemmed Text:")
print(stemmed_words)

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nLemmatized Text:")
print(lemmatized_words)

# Parts of Speech tagging
pos_tags = pos_tag(filtered_words)
print("\nParts of Speech (POS) Tags:")
print(pos_tags)

# Chunking
grammar = r"""
    NP: {<DT>?<JJ>*<NN>}
    VP: {<MD>?<VB.*><NP|PP>}
    PP: {<IN><NP>}
"""
chunk_parser = RegexpParser(grammar)
tree = chunk_parser.parse(pos_tags)
print("\nTree Visualization:")
print(tree)
```

```python
import spacy

# Load the English language model
nlp = spacy.load('en_core_web_sm')

# Moby Dick text
text = """
That night, in the mid-watch, when the old man—as his wont at intervals—stepped forth from the scuttle in which he leaned, and went to his pivot-hole, he suddenly thrust out his face fiercely, snuffing up the sea air as a sagacious ship’s dog will, in drawing nigh to some barbarous isle. He declared that a whale must be near. Soon that peculiar odor, sometimes to a great distance given forth by the living sperm whale, was palpable to all the watch; nor was any mariner surprised when, after inspecting the compass, and then the dog-vane, and then ascertaining the precise bearing of the odor as nearly as possible, Ahab rapidly ordered the ship’s course to be slightly altered, and the sail to be shortened.
"""

# Tokenize the text
doc = nlp(text)

# Filter stopwords and lemmatize
filtered_tokens = [token.lemma_ for token in doc if not token.is_stop]

# Parts of Speech tagging
pos_tags = [(token.text, token.pos_) for token in doc]

# Display the results
print("Tokenized and Lemmatized Text:")
print(filtered_tokens)

print("\nParts of Speech (POS) Tags:")
print(pos_tags)

# Create a displaCy visualization
from spacy import displacy

# Process the text with displaCy
displacy.render(doc, style='dep', options={'distance': 100})
```

```python
from textblob import TextBlob
from textblob import Word

# Reusing the Moby Dick text
text = """
That night, in the mid-watch, when the old man—as his wont at intervals—stepped forth from the scuttle in which he leaned, and went to his pivot-hole, he suddenly thrust out his face fiercely, snuffing up the sea air as a sagacious ship’s dog will, in drawing nigh to some barbarous isle. He declared that a whale must be near. Soon that peculiar odor, sometimes to a great distance given forth by the living sperm whale, was palpable to all the watch; nor was any mariner surprised when, after inspecting the compass, and then the dog-vane, and then ascertaining the precise bearing of the odor as nearly as possible, Ahab rapidly ordered the ship’s course to be slightly altered, and the sail to be shortened.
"""

# Tokenize the text
blob = TextBlob(text)
tokens = blob.words
print("Tokenized Text:")
print(tokens)

# Lemmatization
lemmatized_tokens = [Word(token).lemmatize() for token in tokens]
print("\nLemmatized Text:")
print(lemmatized_tokens)

# Parts of Speech (POS) Tags
pos_tags = blob.tags
print("\nParts of Speech (POS) Tags:")
print(pos_tags)

# Sentiment Analysis
sentiment = blob.sentiment
print("\nSentiment Analysis:")
print(f"Polarity: {sentiment.polarity}, Subjectivity: {sentiment.subjectivity}")
```

```python
class Calculator:
    def addition(self, x, y):
        return x + y

    def subtraction(self, x, y):
        return x - y

    def multiplication(self, x, y):
        return x * y

    def division(self, x, y):
        if y == 0:
            return "Cannot divide by zero"
        return x / y

class Special(Calculator):
    def modulo(self, x, y):
        if y == 0:
            return "Cannot perform modulo with zero"
        return x % y

    def exponential(self, x, y):
        return x ** y
```
